```python
## Player attributes that we want categorized into groupings
player_info = ['_player_id', '_name', '_height', '_weight', '_position']
player_overall = ['_season', '_team_abbreviation', '_minutes_played', '_total_rebounds', '_player_efficiency_rating']
player_offense = ['_field_goal_percentage', '_free_throw_percentage', '_two_point_percentage', '_three_point_percentage', '_offensive_rebounds', '_assists', '_turnovers', '_usage_percentage', '_offensive_win_shares']
player_defense = ['_defensive_rebounds', '_steals', '_blocks', '_defensive_win_shares']

```

```python
## specifically 2018-2019 only
year = '2018-19'
## Seasonal attributes
remaining = player_overall + player_offense + player_defense
individual_player_dict = {}
for player in ucla_players:
    row = []
    
    ## Singlue valued attributes
    for single_attr in player_info:
        item = getattr(player, single_attr)
        row.append(item)
    
    ## find index of specific year
    try:
        year_index = player._season.index(year)
        for single_attr in remaining:
            item = getattr(player, single_attr)[year_index]
            row.append(item)
        individual_player_dict[player._player_id] = row
    except:
        print(player._name, "player didn't play in ", year)
    
```

```python
ucla_df = pd.DataFrame(individual_player_dict).T
```

```python
ucla_df.columns = player_info + remaining
ucla_df = ucla_df.replace('', 0)
ucla_df
```
